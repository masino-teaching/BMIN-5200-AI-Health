```python
import numpy as np
import random
```

```python
class Car:
  def __init__(self, num_doors, num_passengers):
    self.nd = num_doors
    self.np = num_passengers
    self.fuel_economy_rate = 1 + np.random.random()

  def mpg(self, driving_condition):
    if driving_condition == 'highway':
      return 25 * self.fuel_economy_rate
    elif driving_condition == 'city':
      return 10 * self.fuel_economy_rate
    else:
      return 17.5 * self.fuel_economy_rate

  def pretty_print(self):
    print(f"Doors: {self.nd}, Seating: {self.np}, Highway MPG: {self.mpg('highway'):.2f}, City MPG: {self.mpg('city'):.2f}")

def print_cars(cars):
  for c in cars: c.pretty_print()
```

```python
# some built in sorting
l = ['a', 'c', 'b']
l.sort()
print(l)

l = [4, 7, 3]
l.sort()
print(l)
```

```python
my_cars = [Car(2, 2) for _ in range(2)] + [Car(4, 4) for _ in range(2)] + [Car(4, 6) for _ in range(2)]
random.shuffle(my_cars)
print_cars(my_cars)
```

Sort the cars:
1. By number of passengers
2. By highway mpg
3. By city mpg in descending order

```python
my_cars.sort(key=lambda c: c.nd)
print_cars(my_cars)
```

```python
my_cars.sort(key=lambda c: c.mpg('highway'))
print_cars(my_cars)
```

```python
my_cars.sort(key=lambda c: c.mpg('city'), reverse=True)
print_cars(my_cars)
```
